```python

plt.plot(df['Temperature (°C)'],df['Ice Cream Sales (units)'])
plt.show()
```

```python
x= df[['Temperature (°C)']]
x.ndim
y= df['Ice Cream Sales (units)']
```

```python
#Convert data into polynomial function ,then we perdict the x_test vlue
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2)
pf.fit(x)
x=pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
prd=lr.predict(x)
```

```python

plt.plot(df['Temperature (°C)'],df['Ice Cream Sales (units)'],color='blue')
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales (units)')
plt.plot(df['Temperature (°C)'],prd,color='red',linewidth=3)
plt.legend(['Original','Predict'])
plt.show()
```
