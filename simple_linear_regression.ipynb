```python
# x we need two dimension becaused it is independent/input/predictor
x=df[["age"]]
# y we used in single dimension
#becaused We are fitting a model like y=mX+c
#x:2D array (input)=(rows,columns)
#y:1d array(target)=rows
y=df["bought_insurance"]
```

```python
#random_state=42 isaliye used karate hai humara data har bar changr hota
#rahega to change na ho esaliye random_state used=42 ideally number
#we can used any one
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=15)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr= LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```
